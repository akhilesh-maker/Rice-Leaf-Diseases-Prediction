```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

```

```python
dataset_path = 'C:/Users/admin/python/Internship/RiceLeaf/Data' # update path
classes = os.listdir(dataset_path)
print(f"Classes: {classes}")
for cls in classes:
    print(f"{cls}: {len(os.listdir(os.path.join(dataset_path, cls)))} images")

```

```python
fig, axes = plt.subplots(1, len(classes), figsize=(16,5))
for i, cls in enumerate(classes):
    class_path = os.path.join(dataset_path, cls)
    img_path = next(f for f in os.listdir(class_path) if f.lower().endswith(('.jpg','.jpeg','.png')))
    img = Image.open(os.path.join(class_path, img_path))
    axes[i].imshow(img)
    axes[i].set_title(cls)
    axes[i].axis('off')
plt.show()
```
